```python
def single_characters(source_file_name):
    destination_folder_path = r"C:\Users\joesh\Desktop\jupyter\TextFiles_single_removed"

    with open(source_file_name, 'r', encoding="ISO-8859-1") as source_file:
        new_file_name = os.path.basename(source_file_name).replace("unmod", "mod1")
        dest_file_name = os.path.join(destination_folder_path, new_file_name)

        with open(dest_file_name, 'w', encoding="utf-8") as destination_file:
            for line in source_file:
                words = line.split(" ")

                for word in words:
                    if len(word) > 1:
                        word = word.lower()
                        destination_file.write(word + " ")
```

```python
def punctuations(source_file_name):
    destination_folder_path = r"C:\Users\joesh\Desktop\jupyter\TextFiles_punctuations_removed"

    with open(source_file_name, 'r', encoding="ISO-8859-1") as source_file:
        new_file_name = os.path.basename(source_file_name).replace("unmod", "mod")
        dest_file_name = os.path.join(destination_folder_path, new_file_name)

        with open(dest_file_name, 'w', encoding="utf-8") as destination_file:
          text=source_file.readlines()
          words = "\n".join(text)
          
          words = words.lower() # lowercase
          words = word_tokenize(words) # tokenization
          words = [word for word in words if len(word) > 1] 
          words = [word for word in words if word not in stop_words] # non stop words
          words = [word for word in words if word in wordlist] # english words
          words = [' '.join(words[0 : len(words)])]
          words = "\n".join(words)
          destination_file.write(words)
```
